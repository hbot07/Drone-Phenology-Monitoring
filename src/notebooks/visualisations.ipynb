# Side-by-Side Crown Detection Comparison

This notebook visualizes crown detections from two sources:
- **Row 1**: Original input crowns from `input/input_crowns`
- **Row 2**: Output crowns from `output/input_crowns`

```python
# Import required libraries
import geopandas as gpd
import matplotlib.pyplot as plt
import os
from pathlib import Path

# Set paths
input_crowns_dir = Path("../../input/input_crowns")
output_crowns_dir = Path("../../output/input_crowns")

print("Input directory:", input_crowns_dir.resolve())
print("Output directory:", output_crowns_dir.resolve())
```

```python
# List all available GPKG files
input_files = sorted(input_crowns_dir.glob("*.gpkg"))
output_files = sorted(output_crowns_dir.glob("*.gpkg"))

print("Input files:", [f.name for f in input_files])
print("Output files:", [f.name for f in output_files])
```

```python
# Function to create side-by-side comparison
def compare_detections(input_file, output_file, figsize=(16, 8)):
    """
    Create a side-by-side comparison of crown detections
    
    Parameters:
    - input_file: Path to input crown file
    - output_file: Path to output crown file
    - figsize: Figure size (width, height)
    """
    # Load the data
    input_gdf = gpd.read_file(input_file)
    output_gdf = gpd.read_file(output_file)
    
    # Create figure with 2 columns (1 row)
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # Plot input crowns (left)
    input_gdf.plot(ax=axes[0], facecolor='lightgreen', edgecolor='darkgreen', 
                   linewidth=0.5, alpha=0.7)
    axes[0].set_title(f'Input Crowns\n{input_file.name}\n(n={len(input_gdf)})', 
                      fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Longitude')
    axes[0].set_ylabel('Latitude')
    axes[0].grid(True, alpha=0.3)
    
    # Plot output crowns (right)
    output_gdf.plot(ax=axes[1], facecolor='lightcoral', edgecolor='darkred', 
                    linewidth=0.5, alpha=0.7)
    axes[1].set_title(f'Output Crowns\n{output_file.name}\n(n={len(output_gdf)})', 
                      fontsize=14, fontweight='bold')
    axes[1].set_xlabel('Longitude')
    axes[1].set_ylabel('Latitude')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    return fig, axes
```

## Compare All Observation Months (OM1-OM5)

```python
# Create comparisons for all observation months
for input_file, output_file in zip(input_files, output_files):
    print(f"\nComparing {input_file.name}...")
    fig, axes = compare_detections(input_file, output_file)
    plt.show()
```

## Compare Specific Observation Month

You can also compare a specific observation month:

```python
# Example: Compare OM1
om_name = "OM1.gpkg"
input_om1 = input_crowns_dir / om_name
output_om1 = output_crowns_dir / om_name

fig, axes = compare_detections(input_om1, output_om1, figsize=(18, 8))
plt.show()
```

## Create a Multi-Row Grid View (All OMs in one figure)

```python
# Create a comprehensive grid showing all observation months
n_oms = len(input_files)
fig, axes = plt.subplots(n_oms, 2, figsize=(16, 6*n_oms))

for i, (input_file, output_file) in enumerate(zip(input_files, output_files)):
    # Load the data
    input_gdf = gpd.read_file(input_file)
    output_gdf = gpd.read_file(output_file)
    
    # Plot input crowns (left column)
    input_gdf.plot(ax=axes[i, 0], facecolor='lightgreen', edgecolor='darkgreen', 
                   linewidth=0.5, alpha=0.7)
    axes[i, 0].set_title(f'INPUT: {input_file.name} (n={len(input_gdf)})', 
                         fontsize=12, fontweight='bold')
    axes[i, 0].set_xlabel('Longitude')
    axes[i, 0].set_ylabel('Latitude')
    axes[i, 0].grid(True, alpha=0.3)
    
    # Plot output crowns (right column)
    output_gdf.plot(ax=axes[i, 1], facecolor='lightcoral', edgecolor='darkred', 
                    linewidth=0.5, alpha=0.7)
    axes[i, 1].set_title(f'OUTPUT: {output_file.name} (n={len(output_gdf)})', 
                         fontsize=12, fontweight='bold')
    axes[i, 1].set_xlabel('Longitude')
    axes[i, 1].set_ylabel('Latitude')
    axes[i, 1].grid(True, alpha=0.3)

fig.suptitle('Crown Detection Comparison: Input vs Output', 
             fontsize=16, fontweight='bold', y=0.995)
plt.tight_layout()
plt.show()
```
